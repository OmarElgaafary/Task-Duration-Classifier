## 03-01 Importing CSV file and spliting training and test subsets

Before we start training the model using the cleaned dataset, we need to split the dataset into subsets where each subset is going to be used for different purposes being as follows:

- One subset for training the model
- One subset for testing the model

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

jira_df = pd.read_csv("../data/processed/jira_issues_cleaned.csv")

copy_df = jira_df.copy()
copy_df.drop("duration_days")

x = copy_df.drop("duration_category")
y = copy_df["duration_category"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```
